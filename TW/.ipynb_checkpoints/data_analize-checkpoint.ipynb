```python
import pandas as pd
```

```python
data = pd.read_csv("outputData.csv")
data
```

```python
data.groupby(["BUFFER_SIZE", "P/C", "PK_CONFIG",  "RANDOMIZATION_OPTION_NORMAL","FAIR_BUFFER",]).count()
```

```python
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)
def draw_plot(config, data):
    x = []
    y = []
    for i,j in data.items():
        x.append(str(i))
        y.append(j)
    plt.scatter(x,y)
#     plt.figure(figsize=(10,5))
    plt.show()
    
grouped_data = data.groupby(["BUFFER_SIZE", "P/C", "PK_CONFIG",  "RANDOMIZATION_OPTION_NORMAL","FAIR_BUFFER",])
for group,data_for_group in grouped_data:
    print(f"Buffer size is {group[0]}, {group[1]}, number of {group[1]}s is {group[2]}, normal randomization: {group[3]}, fair buffer: {group[4]}")
    min_size = data_for_group['SIZE'].min()
    max_size = data_for_group['SIZE'].max()
    interval = 100
    if(max_size > 5000):
        interval = 1000
    data_for_group['bin'] = pd.cut(data_for_group["SIZE"], [x for x in range(0, max_size+99, interval)])
    a = data_for_group.groupby('bin')["COUNTER"].sum()
    a.plot.box()
    plt.show()
    print()
    draw_plot(group, a)
```
